```python
pr = pd.read_csv('Pull Requests Data.csv', encoding='latin-1')
pr
```

```python
pr.info()
```

```python
pr.drop(['assignee', 'closed_at', 'assignees', 'created_at', 'base', 'comments_url', 'commits_url', 'merge_commit_sha', 'mergeable_state', 'mergeable', 'merged_at', 'merged_by', 'html_url', 'milestone', 'updated_at', 'url', 'user', 'head'], axis=1, inplace=True)
pr
```

```python
pr.set_index('id', inplace=True)
pr
```

```python
pr.query('state == "closed"')
```

```python
pr.groupby('repo_id').number.count().sort_values(ascending=False)
```

```python
pr.describe()
```

```python
pr.query('additions > 10000')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
sns.set_context("poster", font_scale=1, rc={
                "grid.linewidth": 5, "axes.linewidth": 5})

plt.figure(figsize=(20, 10))
ax = sns.barplot(x="number", y="additions", data=pr.query('repo_id == 3297519').sort_values('number').sample(30))
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.plot()
```

```python
plt.figure(figsize=(18, 18))
heatmap = sns.heatmap(pr.corr(), vmin=-1, vmax=1,
                      annot=True, fmt='.2f', annot_kws={"size": 13}, cmap='icefire')
heatmap.set_title('Correlações', pad=20)
plt.show()
```

```python
plt.figure(figsize=(20, 5))
ax = sns.boxplot(pr['additions'])
plt.xticks(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(20, 5))
ax = sns.boxplot(pr['additions'])
plt.xticks(fontsize=14)
ax.set_xlim(-200, 2000)
plt.show()
```
